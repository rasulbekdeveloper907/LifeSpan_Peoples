```python
import os
import json
import time
import pandas as pd
from kaggle.api.kaggle_api_extended import KaggleApi
from tqdm import tqdm  # progress bar uchun

# -----------------------------
# Kaggle API sozlash funksiyasi
# -----------------------------
def setup_kaggle_api():
    kaggle_dir = os.path.join(os.path.expanduser('~'), '.kaggle')
    kaggle_file = os.path.join(kaggle_dir, 'kaggle.json')

    if not os.path.exists(kaggle_dir):
        os.makedirs(kaggle_dir, exist_ok=True)

    if not os.path.exists(kaggle_file):
        print("âš™ï¸ Kaggle API token topilmadi.")
        username = input("Kaggle username kiriting: ").strip()
        key = input("Kaggle API key (token) kiriting: ").strip()
        data = {"username": username, "key": key}
        with open(kaggle_file, 'w', encoding='utf-8') as f:
            json.dump(data, f)
        print(f"âœ… Token saqlandi: {kaggle_file}")

    os.chmod(kaggle_file, 0o600)
    print("ğŸ” Ruxsatlar toâ€˜gâ€˜rilandi.")

    api = KaggleApi()
    api.authenticate()
    print("âœ… Kaggle API muvaffaqiyatli autentifikatsiya qilindi.")

    return api

# -----------------------------
# Unikal datasetlarni olish funksiyasi
# -----------------------------
def fetch_unique_datasets(api, total=50, delay=1.5, max_retries=3):
    all_data = []
    seen_dataset_refs = set()  # Unikal datasetlarni saqlash uchun
    page = 1

    with tqdm(total=total, desc="ğŸ“¦ Datasetlar yigâ€˜ilmoqda") as pbar:
        while len(all_data) < total:
            for attempt in range(max_retries):
                try:
                    datasets = api.dataset_list(sort_by="hottest", page=page)
                    break
                except Exception as e:
                    print(f"âš ï¸ API xatosi: {e}, retry {attempt+1}/{max_retries}")
                    time.sleep(5)
            else:
                print("âŒ API bilan bogâ€˜lanib boâ€˜lmadi, toâ€˜xtatildi.")
                break

            if not datasets:
                print("âŒ Yangi dataset topilmadi, toâ€˜xtatildi.")
                break

            for d in datasets:
                if d.ref not in seen_dataset_refs:
                    seen_dataset_refs.add(d.ref)
                    all_data.append(vars(d))
                    pbar.update(1)
                if len(all_data) >= total:
                    break

            page += 1
            time.sleep(delay)

    print(f"âœ… Jami {len(all_data)} ta unikal dataset yigâ€˜ildi.")
    return all_data

# -----------------------------
# CSV ga saqlash funksiyasi
# -----------------------------
def save_to_csv(data, filename='kaggle_unique_datasets_8000.csv'):
    output_dir = r"C:\Users\Rasulbek907\Desktop\LifeSpan_Peoples\Data"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir, exist_ok=True)
    
    full_path = os.path.join(output_dir, filename)

    df = pd.json_normalize(data)
    df.to_csv(full_path, index=False, encoding='utf-8-sig')
    print(f"ğŸ’¾ Fayl saqlandi: {full_path} ({len(df)} yozuv)")

# -----------------------------
# Main funksiyasi
# -----------------------------
def main():
    api = setup_kaggle_api()
    data = fetch_unique_datasets(api, total=50)  # 8000 dataset uchun
    save_to_csv(data)

# -----------------------------
# Ishga tushirish
# -----------------------------
if __name__ == '__main__':
    main()
    
```
